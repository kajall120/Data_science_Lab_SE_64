```python
num = 10

if num > 0:
    print(f'{num} is a positive number.')
else:
    print(f'{num} is a non-positive number.')

num = -5

if num > 0:
    print(f'{num} is a positive number.')
else:
    print(f'{num} is a non-positive number.')
```
